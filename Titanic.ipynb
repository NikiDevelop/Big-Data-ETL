```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="ticks")
```

```python
titanic = pd.read_csv('train.csv')

titanic.head(30)
```

```python
titanic.set_index('PassengerId').head()
```

```python
# Sumamos los casos de hombres y mujeres
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

# los ponemos en una lista
proportions = [males, females]

# Creamos un gráfico
plt.pie(

    proportions,

    labels = ['Males', 'Females'],

    shadow = False,

    colors = ['orange','black'],

    explode = (0.15 , 0),

    startangle = 90,

    autopct = '%1.1f%%'
    )

plt.axis('equal')

plt.title("Tigres en Malasia")

plt.tight_layout()
plt.show()
```

```python

lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex', fit_reg=False)

lm.set(title = 'Mi titulo')

axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,185)
```

```python
# Los ordenamos
df = titanic.Fare.sort_values(ascending = False)
df

binsVal = np.arange(0,600,10)
binsVal

plt.hist(df, bins = binsVal)

plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Payed Histrogram')

plt.show()
```

```python
import csv

with open('faang_stocks_pandemic_data.csv', newline='', encoding="utf8") as file:
  micsv = csv.reader(file, delimiter=',')
  next(file)
  for lineaTransformada in micsv:

    f = lineaTransformada[1][:-8]
    diff = float(lineaTransformada[2]) - float(lineaTransformada[3])
    c = round(float(lineaTransformada[4]),6) #el round no deja poner 4 decimales, era poner 6 decimales o 2
    v = lineaTransformada[-3][:-4]
    e = lineaTransformada[-1].lower()
    if("facebook" in e.lower()):
      e = "1"
    resultado = [f,c,diff,v,e]
    print(resultado)
```
